**4. What is Prompt Engineering, and explain any two types of prompts used in Generative AI?**
- Two Common Types of Prompts in Generative AI:

**Instruction-Based Prompts**

These prompts give the AI clear directions on what to do.

Example: “Summarize this article in three bullet points.”

Use: Ideal for tasks like summarizing, translating, or explaining because it tells the AI exactly what kind of output is expected.

**Contextual (or Role-Based) Prompts**

These prompts give the AI a role or situation to act within.

Example: “You are a data analyst. Explain what Power BI is in simple language.”

Use: Helps the AI tailor its tone, style, and depth of response to match a specific role or audience.

**8.Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.**


- **Descriptive Analytics – “What happened?”**

What it is: The first step in analytics. It looks at past data to summarize it.

Example: A sales report showing how many products were sold last month.


**Diagnostic Analytics – “Why did it happen?”**

What it is: Goes deeper to find reasons behind past events.

Example: Investigating why sales dropped in a certain region by analyzing marketing campaigns, pricing, or customer feedback.


**Predictive Analytics – “What might happen?”**

What it is: Uses historical data and statistical models to forecast future trends.

Example: Predicting which customers are likely to buy a new product or who might cancel a subscription.


**Prescriptive Analytics – “What should we do?”**

What it is: Recommends actions based on data and predictions.

Example: Suggesting the best marketing strategy to increase sales in a slow region.

**Generative AI Analytics – “Let AI help create solutions”**

What it is: The newest stage, where AI can generate insights, reports, content, or even suggest new ideas.

Example: An AI tool that writes a full marketing report, generates product recommendations, or simulates business scenarios.
